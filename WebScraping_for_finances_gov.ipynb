```python
from bs4 import BeautifulSoup
from urllib.request import urlopen
```

```python
url_finances_presentation = 'http://www.finances.gov.tn/fr/presentation'
url_finances_page2 = 'http://www.finances.gov.tn/fr/la-ministre-des-finances'
url_finances_page3 = 'http://www.finances.gov.tn/fr/conseil-national-de-la-fiscalite-cnf'
url_finances_page4 = 'http://www.finances.gov.tn/fr/conseil-national-de-la-comptabilite-cnc'
url_finances_page5 = 'http://www.finances.gov.tn/fr/apercu-general-sur-la-fiscalite'
url_finances_page6 = 'http://www.finances.gov.tn/fr/cadre-reglementaire-1'
url_finances_page7 = 'http://www.finances.gov.tn/fr/apercu-general-0'
url_finances_page8 = 'http://www.finances.gov.tn/fr/apercu-general-1'
url_finances_page9 = 'http://www.finances.gov.tn/fr/apercu-general'
url_finances_page10 = 'http://www.finances.gov.tn/fr/cadre-reglementaire-4'
url_finances_page11 = 'http://www.finances.gov.tn/fr/cadre-legislatif-et-reglementaire-et-gestion-des-biens-et-avoirs-confisques'
```

```python
response_finances = urlopen(url_finances_presentation)
html_finances = response_finances.read()
html_finances
```

```python
soup_finances = BeautifulSoup(html_finances, 'html.parser')
soup_finances
```

```python
soup_finances.find('div', {'class': "cssPage"}).find('li').get_text()
```

```python
for item in soup_finances.find('div', {'class': "cssPage"}).find('ul').find_all('li'):
    print(item.get_text())
```

```python
for item in soup_finances.find('div', {'class': "cssPage"}).find_all('p'):
    print(item.get_text())
```

```python
for item in soup_finances.find('div', {'class': "cssPage"}).find_all('ul'):
    print(item.get_text())
```

```python
item1 = soup_finances.find('div', {'class': "cssPage"})
for item in item1.find_all('p'):
    print(item.get_text())
    caracteristicas = item1.find('ul')
    
    for caract in caracteristicas.find_all('li'):
        print(caract.get_text())
        
    print('-------------')
```
